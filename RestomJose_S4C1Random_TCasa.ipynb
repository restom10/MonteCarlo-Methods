```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x):
    return np.exp(-x**2)
```

```python
def mifun(x):
    x_0 = 3.0
    a = 0.01
    return np.exp(-(x**2))/((x-x_0)**2 + a**2)
```

```python
def lambd(x,y):
    return (mifun(x)/mifun(y))
```

```python
inicial= np.random.random()*8 -4
```

```python
def distri(pasos,sigma):
    distri=[]
    distri.append(mifun(inicial))
    i=0
    while (i<pasos):
        x=distri[i]+np.random.normal()*sigma
        lam=lambd(x,distri[i])

        if(lam>=1):
            distri.append(x)

        else:
            beta=np.random.random()
            if(lam>=beta):
                distri.append(x)
            else:
                distri.append(distri[i])
        i+=1
        
    return distri
    
```

```python
arr1=distri(100000,5)
arr2=distri(100000,0.2)
arr3=distri(100000,0.01)
arr4=distri(1000,0.1)
arr5=distri(100000,0.1)
```

```python
plt.title('Distri1')
plt.hist(arr1,edgecolor = 'black')
plt.grid(True)
plt.show()
plt.savefig("distri1")
plt.close()
```
