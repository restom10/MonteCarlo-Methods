```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def lanzarPaso(i):
    x_temp= x_crema[i]
    y_temp= y_crema[i]
    x_crema[i]+=np.random.randn()*2.5
    y_crema[i]+=np.random.randn()*2.5
    
    if(math.sqrt((x_crema[i]**2)+(y_crema[i]**2))>math.sqrt(230)):
        x_crema[i]= x_temp
        y_crema[i]= y_temp
        lanzarPaso(i)
            
    
```

```python
radioC= math.sqrt(230)*np.random.rand(10000)
anguloC= 2*np.pi*np.random.rand(10000)
x_circulo= np.sin(anguloC)*radioC
y_circulo= np.cos(anguloC)*radioC

radioCr=math.sqrt(2)*np.random.rand(100)
anguloCr= 2*np.pi*np.random.rand(100)

x_crema= np.sin(anguloCr)*radioCr
y_crema= np.cos(anguloCr)*radioCr
i=0
while (i<10000):
    for n in range(100):
        lanzarPaso(n)      
    i+=1
```

```python
plt.figure() 
plt.scatter(x_circulo,y_circulo)
plt.scatter(x_crema,y_crema,)
plt.xlabel("X") 
plt.ylabel("Y") 
plt.title("circulo")
plt.show()
plt.savefig("cafeCrema.pdf")
plt.close()
```
